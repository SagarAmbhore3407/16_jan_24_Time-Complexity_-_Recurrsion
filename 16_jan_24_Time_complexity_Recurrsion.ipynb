```python
#6. Write a recursive function to calculate the sum of digits of a given positive integer.
#   sum_of_digits(123) -> 6

def sumOfDigits(number):
    if len(str(number)) == 1:
        return number
    return int(str(number)[0])+ sumOfDigits(int(str(number)[1:]))

print(sumOfDigits(1236))

```

```python
#7.  Fibonacci Series
#    Write a recursive function to generate the first n numbers of the Fibonacci series.
#    fibonacci_series(6) -> [0, 1, 1, 2, 3, 5]

def fibonacci_series(num):
    if num <=0:
        return []
    elif num == 1:
        return [0]
    elif num == 2:
        return [0,1]
    else:
        fibonacci = fibonacci_series(num-1)
        fibonacci.append(fibonacci[-1]+fibonacci[-2])
        return fibonacci

fibonacci_series(6)
   
  
```

```python
#9. Given a non-empty string and a dictionary of words, write a recursive function to determine if the string can be 
#   segmented into a space-separated sequence of dictionary words.
#   word_break( leetcode , [ leet , code ]) -> True


def word_break(string,sequence):
    if string in sequence:
        return True
    
    else:
        #Iterating on words from sequence
        for word in sequence:
            #Checking the word in string
            for i in range(len(string)):
                if (i == len(string)-1) or (i == len(word)-1):
                    #Checking remaining
                    if word_break(string[i+1:],sequence):
                        return True           
    return False


print(word_break("leetcodelove",["code","leet","sam","leet","love","adira","sung"]))          
        


```

```python
#10. Implement a recursive function to solve the N Queens problem, where you have to place N queens on an N×N 
#    chessboard in such a way that no two queens threaten each other.
'''  n_queens(4)
        [

         [".Q..",

          "...Q",

          "Q...",

          "..Q."],

         ["..Q.",

          "Q...",

          "...Q",

          ".Q.."]

        ]
'''

#This function checks wheather this passed space is safe or not by comparing with the previous queens.
def checkIsSafePlace(board,row,col,n):
    
    #check backward in same row
    for i in range(col,-1,-1):
        if board[row][i] == "📀":
            return False
        
        
    #check backward upper diagonal
    for r,c in zip(range(row,-1,-1),range(col,-1,-1)):
        if board[r][c] == "📀":
            return False
        
    
    #Check backward lower diagonal
    for r,c in zip(range(row,n,1),range(col,-1,-1)):
        if board[r][c] == "📀":
            return False
        
    return True


#This is Recurrsive func which fills the queens in appropriate space
def queensUtil(board,col,n):
    if col>=n:
        return True
    
    else:
        for row in range(n):
            if checkIsSafePlace(board,row,col,n):
                board[row][col]="📀"
                
                #Check for another column , if successful return True otherwise False
                if queensUtil(board,col+1,n):
                    return True
                
                # If false then the queen again removed from same space
                else:
                    board[row][col]="💿"
                
        return False

      

#This func is used to print Chessboard
def printBoard(board,n):
    for row in range(n):
        for col in range(n):
            print(f"|{board[row][col]}",end="")
        print("|")     


#This function used to call the recurrsive fun, and also prints the board by calling above func        
def fillQueens(board,n,col = 0):
    if not queensUtil(board,col,n):
        print("Cannot fills")
        return
    printBoard(board,n)
       
        
        
n = 8     
board = [["💿","💿","💿","💿","💿","💿","💿","💿"],
         ["💿","💿","💿","💿","💿","💿","💿","💿"],
         ["💿","💿","💿","💿","💿","💿","💿","💿"],
         ["💿","💿","💿","💿","💿","💿","💿","💿"],
         ["💿","💿","💿","💿","💿","💿","💿","💿"],
         ["💿","💿","💿","💿","💿","💿","💿","💿"],
         ["💿","💿","💿","💿","💿","💿","💿","💿"],
         ["💿","💿","💿","💿","💿","💿","💿","💿"]]

print(f":: The {n}*{n} Queen Chess-Board ::")
print(f"\nQueen Occupied ->📀\nVacant Space ->💿\n\nBefore:  ")
printBoard(board,n)
print("\nAfter:  ")
fillQueens(board,n)


        

```
